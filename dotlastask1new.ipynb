```python
import pandas as pd
from selenium import webdriver
driver = webdriver.Chrome()
from bs4 import BeautifulSoup as bs
import requests
import csv
from lxml import html
```

```python
driver.get('https://www.talabat.com/uae/restaurant/621133/ginos-deli-jlt?aid=1308')
```

```python
url = 'https://www.talabat.com/uae/restaurant/621133/ginos-deli-jlt?aid=1308'
r = requests.get('https://www.talabat.com/uae/restaurant/621133/ginos-deli-jlt?aid=1308')
soup = bs(r.content,'lxml')
latlong = requests.get('https://www.latlong.net/place/al-ain-united-arab-emirates-2522.html')
soup2 = bs(latlong.content,'lxml')
```

```python
lat_long = soup2.find('p').text.split()
lati = lat_long[5]
lati = lati.replace(',','')
lati = float(lati)
print(lati)
longi = lat_long[6]
longi = longi.removesuffix('.')
longi = float(longi)
print(longi)
```

```python
header = soup.find('h1', attrs={'class':'f-24 f-m-15 media-heading f-400 mb-0'})
header_name = header.getText()
print(header_name)

logo = soup.find('img', attrs={'alt':'Gino\'s Deli'})
logo_url = logo['src']
print(logo_url)
```

```python
cuisine_tags = []
menu_items = []
menu_details = []
```

```python
cuisine = soup.find('p',attrs={'class':'light-text f-14 mb-0'}).text
cuisine_tags = []
cuisine_tags = cuisine.split(', ')
print(cuisine_tags)
```

```python
dishes = driver.find_elements_by_class_name('menu-itemstyles__MenuItemContainer-sc-wyx6cl-0')
menu_details.clear()
for dish in dishes:
    name = dish.find_element_by_class_name('f-15').text
    desc = dish.find_element_by_class_name('f-12').text
    price= float(dish.find_element_by_class_name('currency').text)
    #img_class=dish.find_element_by_class_name('lazyImagestyles__FadeContainer-sc-cl9o9m-1')
    #img_url=img_class.find_element_by_css_selector('img').get_attribute('src')
    print(name)
    #print(desc)
    #print(price)
    #print(img_url)
    menu_items.append(name)
    menu_item = {
        'item_name': name,
        'item_description': desc,
        'item-price': price,
        #'item-image': img_url
    }
    menu_details.append(menu_item)

#print(menu_items)
```

```python
df = pd.DataFrame(menu_details)
df.to_csv('menu1.csv')
df
```

```python
csv_file = open('cms_scrape.csv', 'w')
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['restaurant_name', 'restaurant_logo', 'latitude','longitude','cuisine_tags','menu_items'])
```

```python
csv_writer.writerow([header_name,logo_url,lati,longi,cuisine_tags,menu_items])
```
